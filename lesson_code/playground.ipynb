```python
import matplotlib.pyplot as plt
import numpy as np

n = 100
rng = np.random.default_rng(1234)
beta_0 = -4
beta_1 = 1

length = rng.uniform(5, 60, size=n)

log_weight_mu = beta_0 + beta_1 * np.log(length)
log_weight = log_weight_mu + rng.normal(scale=0.5, size=n)
weight = np.exp(log_weight)

x = np.linspace(0.0001, np.max(length), num=100)
y = np.exp(beta_0 + beta_1 * np.log(x))

fig, ax = plt.subplots(figsize=(7, 5))

ax.scatter(length, weight, alpha=0.6)
ax.plot(x, y, color="C1");
```

```python
x = np.linspace(-10, 10, num=200)
y = 1.01 ** x
plt.plot(x, y);
```

```python
x = np.linspace(-10, 10, num=200)
y = 100 * (1.01 ** x) - 100

plt.plot(x, y, label="1.01 ** x");
plt.plot(x, x, ls="--", color="0.5", label="y = x")
plt.legend();
```
